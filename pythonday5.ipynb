Filtering Data (Conditional Statement)

```python
#mengambil sebagian data (terdapat conditional statement)
#mengambil data sales yang lebih dari 500
#head(10) mengambil 10 data teratas
df[df['sales'] > 500].head(10)
```

```python
#menggunakan query style
df.query('sales > 500').head(10)
```

```python
#multiple condition
# & untuk dan
# | untuk or
# ~ untuk not
df[(df['sales'] > 500) & (df['quantity'] > 10)].head()
```

```python
#multiple condition dengan query style
df.query('sales > 500 & quantity > 10').head()
```

```python
df.query('sales <= 500 & quantity <= 10')
```

```python
#menggunakan negasi
df[~(df['sales'] > 500) & (df['quantity'] > 10)].head()
```

Sorting (mengurutkan)

```python
#mengurutkan kolom sales dari kecil ke besar (ascending) = true
df.sort_values('sales', ascending = True)
```

```python
#secara otomatis, python mengurutkan secara ascending
#descending bisa digunakan dengan cara ascensing = False
df.sort_values('sales', ascending = False)
```

```python
#positional argument
def pengurangan(x, y):
  return x - y

pengurangan(2,1)
```

```python
pengurangan(1,2)
```

```python
#melakukan sorting pada beberapa kolom
df.sort_values(['sales', 'quantity'], ascending = [True, False])
```

Value counts: melihat seberapa sering data muncul

```python
df['region'].value_counts()
```

```python
#melihat dalam bentuk persentase
#lebih direkomendasikan untuk melihat frekuensi data muncul pada tabel
df['region'].value_counts(normalize=True)*100
```

```python
#dibelakang koma ada 2
#np.round(values, 2)
np.round(df['region'].value_counts(normalize=True)*100, 2)
```

Summarizing Data

```python
#group by order_date, region, category dengan hasil sales dan quantity sudah dijumlahkan sesuai group
summarize_df = df.groupby(['order_date', 'region', 'category']).agg({'sales': 'sum', 'quantity': 'sum'})
summarize_df
```
