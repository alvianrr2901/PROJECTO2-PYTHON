DICTIONARY

```python
pegawai = {
    'Nama' : 'Dicky',
    'Usia' : '30',
    'Role' : 'Data Analyst',
    'Dept' : 'Marketing'
}
```

```python
pegawai['Nama']
```

```python
#merubah data
pegawai['Nama'] = 'Ricky'
pegawai
```

```python
#menambahkan data ke dalam dictionary
pegawai['Gaji'] = 10000000
pegawai
```

```python
#menghabus data di dictionary
del pegawai['Gaji']
pegawai
```

```python
pegawai.keys()
```

```python
pegawai.values()
```

```python
pegawai.items()
```

```python
for key, value in pegawai.items():
  print(f'{key} : {value}')
```

```python
pegawai['Gaji'] = 12000000
```

```python
pegawai
```

```python
for key, value in pegawai.items():
  print(f'{key} : {value}')
```
