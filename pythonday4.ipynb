enumerate: digunakan untuk memberikan nomor

```python
list_name = ['asep', 'amir', 'budi', 'cinta', 'edi', 'asep']

asep = []
for getIndex, getName in enumerate(list_name):
  if getName == 'asep':
    asep.append(getIndex)
print(asep)
```

```python
asep = [getName for getIndex, getName in enumerate(list_name) if getName == 'asep']
print(asep)
```

```python
bilangan = [2, 4, 3, 6, 7, 9]
bilangan_baru = []

for i in bilangan:
  if i % 2 == 0:
    bilangan_baru.append(i + 1)
  else:
    bilangan_baru.append(i + 2)
print(bilangan_baru)
```

```python
bilangan = [2, 4, 3, 6, 7, 9]

def ganjil_genap(angka):
  bilangan_baru = []

  for i in angka:
    if i % 2 == 0:
      bilangan_baru.append(i+1)
    else:
      bilangan_baru.append(i+2)
  return bilangan_baru

print(ganjil_genap(bilangan))
```

```python
bilangan = [2, 4, 3, 6, 7, 9]
def ganjil_genap(angka):
  if angka % 2 == 0:
    return angka+1
  else:
    return angka+2

bilangan_baru = list(map(ganjil_genap, bilangan))
print(bilangan_baru)
```
